```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_diabetes = pd.read_csv('diabetes_model.csv')
```

```python
data_diabetes.head()
```

```python
data_diabetes.isna().any()
```

##  1. Create a Count plot of Outcome

```python
sns.countplot(data_diabetes['Outcome'])
```

## Analyze the effect of age on Outcome using Boxplot

```python
sns.boxplot(x='Outcome',y='Age',data = data_diabetes,)
```

## Check for the chances of getting diabetes if it is common in family, using the Daiabetes Pedigree Function

```python
sns.boxplot(x='Outcome',y='DiabetesPedigreeFunction',data = data_diabetes)
```

## Check for the relationship with Insulin & Glucose and group the datapoints by Outcome.

```python
sns.scatterplot(data_diabetes['Insulin'], data_diabetes['Glucose'],hue=data_diabetes['Outcome'] )
```

## Create a heatmap on Correlation of features in the data.

```python
fig,ax = plt.subplots(figsize=(8,8))
sns.heatmap(data_diabetes.corr(),annot=True,ax = ax)
```
